```python
def convertairline_sentiment(word):
    if word=='negative':
        return 0
    elif word=='neutral':
        return 1
    else: 
        return 2
    
```

```python
train['airline_sentiment']=train['airline_sentiment'].apply(convertairline_sentiment)
```

```python
from nltk.corpus import stopwords
stop=stopwords.words('english')
import string
import re
punctuations=list(string.punctuation)
stop=stop + punctuations
```

```python
def cleantweet(word):
    letters_only = re.sub("[^a-zA-Z]", " ",word)
    words=letters_only.split()
    cleanlist=[w.lower() for w in words if not w in stop]
    return( " ".join(cleanlist))
```

```python
train['tweet']=train['text'].apply(lambda x:cleantweet(x))
```
